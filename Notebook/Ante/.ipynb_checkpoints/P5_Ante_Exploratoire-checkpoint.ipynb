# Creating a Dataset #

To ease the futur segmentation process, we will create a unique dataset with all the necessary informations.

## Choosing the Variables ##

Basing our research on the main types of segmentation (RFM and Behavior), we will select variables from the existings dataset and creating some new variables from the available informations.

## Creating the dataset ## 

### Retrieving the existing informations ###

We will firstly retrieve the variables that we will keep as it is. 

```python
data_df = pd.DataFrame()
data_df['Customer ID'] = customers_df['customer_unique_id']
data_df['Customer City'] = customers_df['customer_city']
```

## Creating New Variables ##

To ease this process, we will first attach the orders' ID to the corresponding customer.

```python
for customer in data_df['Customer ID']:
    index = data_df.loc[data_df['Customer ID'] == customer].index[0]
    ID_list = list(customers_df[customers_df["customer_unique_id"] == customer]['customer_id'])
    order_list = []
    for ID in ID_list:
        order_list.append(list(orders_df[orders_df["customer_id"] == ID]['order_id'])[0])
    if type(order_list) != list:
        data_df.loc[index, 'Orders Made'] = list(order_list)
    else:
        data_df.loc[index, 'Orders Made'] = order_list
    
# A small inconsistency fixed
temporary_list = []
temporary_list.append(data_df.loc[0,'Orders Made'])
data_df.loc[0,'Orders Made'] = temporary_list
```

Now we will look for the most recent order of the customers.

```python
# We convert the column from string to date.
orders_df['order_purchase_timestamp'] = pd.to_datetime(orders_df.order_purchase_timestamp)

# We fill our new dataset.
for i in range(len(data_df)):
    order_list = data_df.loc[i, 'Orders Made']
    order_dates = []
    if type(order_list) == list:
        for order in order_list:
            order_dates.append(list(orders_df[orders_df["order_id"] == order]['order_purchase_timestamp'])[0])
    if len(order_dates) != 0:
        data_df.loc[i, 'Last Order Date'] = max(order_dates)
```

We will count the total number of orders per customer.

```python
for i in range(len(data_df)):
    if type(data_df.loc[i,'Orders Made']) == float:
        data_df.loc[i,'Total Number of Orders'] = 0
    else:
        data_df.loc[i,'Total Number of Orders'] = len(data_df.loc[i,'Orders Made'])
```

From there, we will look at the average number of order per trimester and the number of order last trimester.
The first purchase we have data on was done on 2016-09-04 and the last on 2018-10-17. Though, we have data on 10 trimester.

```python
for i in range(len(data_df)):
    data_df.loc[i,'Average Number of Orders per Trimester'] = data_df.loc[i,'Total Number of Orders']/10
    order_list = data_df.loc[i, 'Orders Made']
    last_trimester_order = 0
    if type(order_list) == list:
        for order in order_list:
            order_date = list(orders_df[orders_df["order_id"] == order]['order_purchase_timestamp'])[0]
            if order_date >= pd.to_datetime('2018-10-01 00:00:00'):
                last_trimester_order += 1
    data_df.loc[i, 'Number of Orders Last Trimester'] = last_trimester_order
```

We will now look at the average number of items per order.

```python
for i in range(len(data_df)):
    order_list = data_df.loc[i, 'Orders Made']
    items_number_list = []
    if type(order_list) == list:
        for order in order_list:
            order_items = []
            order_items = list(items_df[items_df["order_id"] == order]['order_item_id'])
            if len(order_items) != 0:
                items_number_list.append(max(order_items))
    data_df.loc[i, 'Average Number of Items per Order'] = np.mean(items_number_list)
```

We will now get different informations on payment.

```python
for i in range(len(data_df)):
    order_list = data_df.loc[i, 'Orders Made']
    total_payment = 0
    last_trimester_payment = 0
    payment_installments = []
    if type(order_list) == list:
        for order in order_list:
            if order in list(payments_df["order_id"]):
                payment = list(payments_df[payments_df["order_id"] == order]['payment_value'])[0]
                installments = list(payments_df[payments_df["order_id"] == order]['payment_installments'])[0]
                payment_installments.append(installments)
                total_payment += payment
                if order_date >= pd.to_datetime('2018-10-01 00:00:00'):
                    last_trimester_payment += payment
    data_df.loc[i, 'Total Payment'] = total_payment
    data_df.loc[i, 'Average Payment per Trimester'] = total_payment/10
    data_df.loc[i, 'Payment Last Trimester'] = last_trimester_payment
    data_df.loc[i, 'Average Payment per Order'] = total_payment/data_df.loc[i,'Total Number of Orders']
    if len(payment_installments) != 0:
        data_df.loc[i, 'Average Payment Installments'] = np.mean(payment_installments)
```

Next up, we will found outhow much orders have been made in the same city and how many different category of product have been purchased.

```python
for i in range(len(data_df)):
    order_list = data_df.loc[i, 'Orders Made']
    product_category_list = []
    product_id = []
    same_city = 0
    if type(order_list) == list:
        for order in order_list:
            if order in list(items_df['order_id']):
                seller_id = list(items_df[items_df["order_id"] == order]['seller_id'])[0]
                seller_city = list(sellers_df[sellers_df['seller_id'] == seller_id]['seller_city'])[0]
                order_products = list(items_df[items_df["order_id"] == order]['product_id'])
                for product in order_products:
                    product_id.append(product)
                if data_df.loc[i, 'Customer City'] == seller_city:
                    same_city += 1
    for product in product_id:
        product_category = list(products_df[products_df["product_id"] == product]['product_category_name'])[0]
        if product_category not in product_category_list:
            product_category_list.append(product_category)
    if same_city != 0:
        percent = same_city/data_df.loc[i,'Total Number of Orders']*100
        data_df.loc[i,'Percentage of Orders in the Same City'] = percent
    if same_city == 0:
        data_df.loc[i,'Percentage of Orders in the Same City'] = 0
    data_df.loc[i, 'Number of Different Product Categories'] = len(product_category_list)
```

At last, we will take a look at different variables around the reviews and score.

```python
# From string to date format
orders_df['order_delivered_customer_date'] = pd.to_datetime(orders_df.order_purchase_timestamp)
reviews_df['review_creation_date'] = pd.to_datetime(reviews_df.review_creation_date)

# Filling up the dataset
for i in range(len(data_df)):
    order_list = data_df.loc[i, 'Orders Made']
    review_list = []
    time_to_review_list = []
    score_list = []
    if type(order_list) == list:
        for order in order_list:
            reception_date = np.nan
            time_to_review = np.nan
            if order in list(orders_df['order_id']):
                reception_date = list(orders_df[orders_df['order_id'] == order]['order_delivered_customer_date'])[0]
            if order in list(reviews_df['order_id']):
                review_id = list(reviews_df[reviews_df['order_id'] == order]['review_id'])[0]
                review_score = list(reviews_df[reviews_df['order_id'] == order]['review_score'])[0]
                review_date = list(reviews_df[reviews_df['order_id'] == order]['review_creation_date'])[0]
                review_list.append(review_id)
                score_list.append(review_score)
                if type(reception_date) != float:
                    time_to_review = review_date - reception_date
                if type(time_to_review) != float:
                    time_to_review_list.append(time_to_review)
    data_df.loc[i, 'Number or Reviews Given'] = len(review_list)
    if data_df.loc[i, 'Total Number of Orders'] != 0:
        data_df.loc[i, 'Percentage of Reviews Given'] = len(review_list)/data_df.loc[i, 'Total Number of Orders'] * 100 
    if len(time_to_review_list) != 0:
        data_df.loc[i, 'Average Time to Review (days)'] = np.mean(time_to_review_list)
    if len(score_list) != 0:
        data_df.loc[i, 'Average Score Given'] = np.mean(score_list)
```

```python
data_df.to_csv('Ressources' + '/' + 'TempData.csv', index=False)
```
